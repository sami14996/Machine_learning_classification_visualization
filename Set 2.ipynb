```python
import sklearn
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
```

# DATASET = CIRCLES0.3

```python
dataset1 = pd.read_csv('C:\Python37\datasets\circles0.3.csv')
```

```python
df1=pd.DataFrame(dataset1)
df1_train, df1_test = train_test_split(df1,test_size=0.3)# hold out 70,30
df1x_train=df1_train.loc[:,['x1','x2']]
df1y_train=df1_train.label
df1x_test=df1_test.loc[:,['x1','x2']]
df1y_test=df1_test.label
```

```python
colors=['red' if label==0 else 'orange' for label in df1.label]
df1_1 = df1.loc[df1.label==1]
df1_0 = df1.loc[df1.label==0]
plt.scatter(df1_1.x1,df1_1.x2,c='red',marker='2')
plt.scatter(df1_0.x1,df1_0.x2,c='blue',marker='+')
plt.show()
```

```python
model = LinearDiscriminantAnalysis()
model.fit(df1x_train,df1y_train)
df1_pred = model.predict(df1x_test)
print("Accuracy",model.score(df1x_test,df1y_test))
confusion_matrix = sklearn.metrics.confusion_matrix(df1y_test,df1_pred)
tp = confusion_matrix[0][0]
fn = confusion_matrix[0][1]
fp = confusion_matrix[1][0]
tn = confusion_matrix[1][1]
print("Confusion Matrix\n",confusion_matrix)
print("PPV",tp/(tp+fp))
print("NPV",tn/(tn+fn))
print("Specificity",tn/(tn+fp))
print("Sensitivity",tp/(tp+fn))
```

```python
model = QuadraticDiscriminantAnalysis()
model.fit(df1x_train,df1y_train)
df1_pred = model.predict(df1x_test)
print("Accuracy",model.score(df1x_test,df1y_test))
confusion_matrix = sklearn.metrics.confusion_matrix(df1y_test,df1_pred)
tp = confusion_matrix[0][0]
fn = confusion_matrix[0][1]
fp = confusion_matrix[1][0]
tn = confusion_matrix[1][1]
print("Confusion Matrix\n",confusion_matrix)
print("PPV",tp/(tp+fp))
print("NPV",tn/(tn+fn))
print("Specificity",tn/(tn+fp))
print("Sensitivity",tp/(tp+fn))
```

```python
model = GaussianNB()
model.fit(df1x_train,df1y_train)
df1_pred = model.predict(df1x_test)
print("Accuracy",model.score(df1x_test,df1y_test))
confusion_matrix = sklearn.metrics.confusion_matrix(df1y_test,df1_pred)
tp = confusion_matrix[0][0]
fn = confusion_matrix[0][1]
fp = confusion_matrix[1][0]
tn = confusion_matrix[1][1]
print("Confusion Matrix\n",confusion_matrix)
print("PPV",tp/(tp+fp))
print("NPV",tn/(tn+fn))
print("Specificity",tn/(tn+fp))
print("Sensitivity",tp/(tp+fn))
```

```python
model = svm.SVC()#Default Kernel RBF
model.fit(df1x_train,df1y_train)
df1_pred = model.predict(df1x_test)
print("Accuracy",model.score(df1x_test,df1y_test))
confusion_matrix = sklearn.metrics.confusion_matrix(df1y_test,df1_pred)
tp = confusion_matrix[0][0]
fn = confusion_matrix[0][1]
fp = confusion_matrix[1][0]
tn = confusion_matrix[1][1]
print("Confusion Matrix\n",confusion_matrix)
print("PPV",tp/(tp+fp))
print("NPV",tn/(tn+fn))
print("Specificity",tn/(tn+fp))
print("Sensitivity",tp/(tp+fn))
```
